```python
q = QuantumRegister(1)
circ = QuantumCircuit(q)


%matplotlib inline

##circuit.data lists all existing elements
##circuit.data.pop(n) gets rid of existing gates you don't want
##for i in range(0,n):
##    circuit.data.pop(i)

circuit.barrier()
circuit.measure([0,1],[0,1])

circuit.draw(output = 'mpl')
```

```python
circuit.barrier()
circuit.cx(1,2)
circuit.cz(0,2)
circuit.barrier()
circuit.draw(output = 'mpl')
```

```python
circuit.measure(2,2)
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend = simulator, shots = 1024).result()
counts = result.get_counts()
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```
